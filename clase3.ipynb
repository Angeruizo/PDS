```python
##SEÑAL CONTINUA

##Importa librerias

import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

##Crea un vector de muestras de -5 a 5 segundos con un paso de 0.01  segundos
t= np.arange(-5,5,0.01)
x=t**2

##Graficar la señal:
plt.plot(t,x,label='Señal x= t^2')
plt.xlabel('Tiempo(s)')
plt.ylabel('Amplitud')
plt.title('Señal x=t^2')
plt.grid(True)
plt.show()

##SEÑAL DISCRETA
n= np.arange(-5,6,1)
y=n**2

#Graficar la señal

plt.stem(n,y,label='Señal x=n^2')
plt.xlabel('Muestra(n)')
plt.ylabel('Amplitud')
plt.title('Señal x=n^2')
plt.grid(True)
plt.show()

## Se puede discretizar o tener una información temporal al graficar nTs
```
